Definicion del método comparar números, en el que se muestra el empleo de PEP8. En este se muestra como documentar un método y como nombrar tambien. Empleando una llamada para probar que el codigo esta funcionando

```python
def compare_numbers(a, b):
    """
    Compare two numbers and return the larger number.

    Parameters:
        a (int or float): First number to be compared.
        b (int or float): Second number to be compared.

    Returns: The larger number.
    """
    if a > b:
        result = a
    else:
        result = b
    return result


if __name__ == "__main__":
    # Llamada al metodo compare_numbers()
    larger_number = compare_numbers(21,15)
    print(larger_number)
```

Context Manager
Context managers are objects that define the __enter__() and __exit__() methods. These methods allow you to use the object as a context manager using the with statement. Here's an example of a custom context manager that show's how could be a file manager:


```python
class FileManager:
    """
    Context manager for handling files.

    Opens a file when entering the context.

    Args:
        filename (str): The path of the file to open.
        mode (str): The mode in which to open the file.
    """

    def __init__(self, filename, mode):
        self.filename = filename
        self.mode = mode

    def __enter__(self):
        """
        Open the file.

        Returns:
            file object: The opened file object.
        """
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_val, exc_tb):
        """
        Close the file.
        """
        self.file.close()


with FileManager("example.txt", "w") as f:
    # Code goes here
    pass
```

Mostrar mensaje aplicando style code pep8 en python empleando metodo delete_vowels.

```python
def delete_vowels(string):
    """
    Delete all the vowels in a string.

    Args:
        string: String from which vowels will be deleted.

    Returns:
        A string without vowels.
    """
    vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
    result = "".join(char for char in string if char not in vowels)
    return result


if __name__ == "__main__":
    string = input("Enter a string: ")
    result = delete_vowels(string)
    print(result)
```
